# Converting all the extensions of images to jpg

```python
import os
import cv2

source_dir = "/content/images"

# Loop through all files in the folder
for filename in os.listdir(source_dir):
    name, ext = os.path.splitext(filename)
    ext = ext.lower()

    if ext in ['.jpeg', '.jpg', '.png', '.jfif']:
        img_path = os.path.join(source_dir, filename)
        img = cv2.imread(img_path)

        if img is not None:
            new_name = name + ".jpg"
            new_path = os.path.join(source_dir, new_name)

            # Save image as .jpg (overwrite if already .jpg)
            cv2.imwrite(new_path, img)

            # Delete original file if different format
            if ext != ".jpg":
                os.remove(img_path)
        else:
            print(f"⚠️ Skipped unreadable image: {filename}")

print("✅ All images converted to .jpg in:", source_dir)
```

# Renaming all the images and its annotations for easier tracking

```python
import os
import shutil

# Original paths
image_dir = "/content/images"
anno_dir = "/content/annotations"

# Target folders for renamed files
ren_image_dir = "/content/ren_images"
ren_anno_dir = "/content/ren_annotations"

# Create target folders if they don't exist
os.makedirs(ren_image_dir, exist_ok=True)
os.makedirs(ren_anno_dir, exist_ok=True)

# Get all .jpg files
image_files = sorted([f for f in os.listdir(image_dir) if f.lower().endswith(".jpg")])

# Counter for naming
count = 1

for img_file in image_files:
    base_name, _ = os.path.splitext(img_file)

    # Check for annotation (either .xml or .txt)
    xml_file = base_name + ".xml"
    txt_file = base_name + ".txt"

    # Define new base name
    new_base = f"img_{count:05d}"

    # Rename + move image
    old_img_path = os.path.join(image_dir, img_file)
    new_img_path = os.path.join(ren_image_dir, new_base + ".jpg")
    shutil.move(old_img_path, new_img_path)

    # Rename + move annotation
    if os.path.exists(os.path.join(anno_dir, xml_file)):
        shutil.move(
            os.path.join(anno_dir, xml_file),
            os.path.join(ren_anno_dir, new_base + ".xml")
        )
    elif os.path.exists(os.path.join(anno_dir, txt_file)):
        shutil.move(
            os.path.join(anno_dir, txt_file),
            os.path.join(ren_anno_dir, new_base + ".txt")
        )
    else:
        print(f"⚠️ No annotation found for: {img_file} — deleting image")
        os.remove(new_img_path)
        continue  # skip incrementing if file is deleted

    count += 1

print("✅ All images and annotations renamed & moved successfully.")
```

# Making a single csv for Labels

```python
import os
import xml.etree.ElementTree as ET
import pandas as pd
from PIL import Image

# Paths
image_dir = "/content/images"
ann_dir = "/content/annotations"

data = []

# --- 1. Process XML (Pascal VOC)
for xml_file in os.listdir(ann_dir):
    if not xml_file.endswith(".xml"):
        continue

    try:
        tree = ET.parse(os.path.join(ann_dir, xml_file))
        root = tree.getroot()

        # Force filename to match XML filename
        base_name = os.path.splitext(xml_file)[0]
        for ext in [".jpg", ".jpeg", ".png"]:
            image_path = os.path.join(image_dir, base_name + ext)
            if os.path.exists(image_path):
                filename = base_name + ext
                break
        else:
            continue  # Skip if image not found

        for obj in root.findall("object"):
            bbox = obj.find("bndbox")
            if bbox is None:
                continue
            xmin = int(float(bbox.find("xmin").text))
            ymin = int(float(bbox.find("ymin").text))
            xmax = int(float(bbox.find("xmax").text))
            ymax = int(float(bbox.find("ymax").text))
            data.append([filename, xmin, ymin, xmax, ymax])

    except Exception as e:
        print(f"❌ Error parsing {xml_file}: {e}")
        continue

# --- 2. Process YOLO TXT
for txt_file in os.listdir(ann_dir):
    if not txt_file.endswith(".txt"):
        continue

    base_name = os.path.splitext(txt_file)[0]

    for ext in [".jpg", ".jpeg", ".png"]:
        image_path = os.path.join(image_dir, base_name + ext)
        if os.path.exists(image_path):
            filename = base_name + ext
            break
    else:
        continue  # No matching image found

    try:
        img = Image.open(image_path)
        img_w, img_h = img.size

        with open(os.path.join(ann_dir, txt_file), 'r') as f:
            lines = f.readlines()
            for line in lines:
                parts = line.strip().split()
                if len(parts) != 5:
                    continue
                class_id, x_center, y_center, width, height = map(float, parts)
                x_center *= img_w
                y_center *= img_h
                width *= img_w
                height *= img_h

                xmin = int(x_center - width / 2)
                ymin = int(y_center - height / 2)
                xmax = int(x_center + width / 2)
                ymax = int(y_center + height / 2)

                data.append([filename, xmin, ymin, xmax, ymax])

    except Exception as e:
        print(f"❌ Error processing {txt_file}: {e}")
        continue

# --- 3. Save to CSV
df = pd.DataFrame(data, columns=["filename", "xmin", "ymin", "xmax", "ymax"])
df.to_csv("/content/final_annotations.csv", index=False)

print("✅ Final CSV created: /content/final_annotations.csv")
```

# Data Preprocessing

```python
import cv2
import os
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataframe
df = pd.read_csv("/content/final_annotations.csv")

# Directory where images are stored
image_dir = "/content/images"

# Folder to save annotated images
save_dir = "/content/annotated_images"
os.makedirs(save_dir, exist_ok=True)

# Function to annotate a single image using bounding boxes
def annotate_image(filename, bboxes):
    img_path = os.path.join(image_dir, filename)
    img = cv2.imread(img_path)

    if img is None:
        print(f"⚠️ Image not found or unreadable: {filename}")
        return

    # Draw bounding boxes
    for xmin, ymin, xmax, ymax in bboxes:
        cv2.rectangle(img, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)
        cv2.putText(img, "Plate", (xmin, ymin - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

    # Save annotated image
    save_path = os.path.join(save_dir, filename)
    cv2.imwrite(save_path, img)

# Annotate ALL images in the dataframe
unique_filenames = df['filename'].unique()

for filename in unique_filenames:
    bboxes = df[df['filename'] == filename][['xmin', 'ymin', 'xmax', 'ymax']].values
    annotate_image(filename, bboxes)

print("✅ All annotated images saved in:", save_dir)
```

```python
import cv2
import numpy as np
import os
from tqdm import tqdm

# Input and Output Directories
input_dir = "/content/images"
processed_dir = "/content/processed"
grayscale_dir = "/content/grayscale"
edges_dir = "/content/edges"

# Create output folders if they don't exist
os.makedirs(processed_dir, exist_ok=True)
os.makedirs(grayscale_dir, exist_ok=True)
os.makedirs(edges_dir, exist_ok=True)

# Resize with padding function
def resize_with_padding(image, target_size=(224, 224), pad_color=(0, 0, 0)):
    h, w = image.shape[:2]
    target_w, target_h = target_size

    scale = min(target_w / w, target_h / h)
    new_w = int(w * scale)
    new_h = int(h * scale)

    resized = cv2.resize(image, (new_w, new_h), interpolation=cv2.INTER_AREA)

    top = (target_h - new_h) // 2
    bottom = target_h - new_h - top
    left = (target_w - new_w) // 2
    right = target_w - new_w - left

    padded = cv2.copyMakeBorder(resized, top, bottom, left, right, cv2.BORDER_CONSTANT, value=pad_color)
    return padded

# Process all images in the input folder
image_files = [f for f in os.listdir(input_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

for filename in tqdm(image_files, desc="🔄 Processing images"):
    image_path = os.path.join(input_dir, filename)
    img = cv2.imread(image_path)

    if img is None:
        print(f"⚠️ Could not read image: {filename}")
        continue

    # Resize + pad
    resized_img = resize_with_padding(img, target_size=(224, 224))

    # Grayscale + equalize
    gray = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY)
    gray_eq = cv2.equalizeHist(gray)

    # Canny edge detection
    edges = cv2.Canny(gray_eq, 50, 150)

    # Save images
    cv2.imwrite(os.path.join(processed_dir, filename), resized_img)
    cv2.imwrite(os.path.join(grayscale_dir, filename), gray_eq)
    cv2.imwrite(os.path.join(edges_dir, filename), edges)

print("✅ All images processed and saved in respective folders.")
```

# CNN

```python
# 1st approach - 3 → 32 → 32 → 64 → 128 → 128 → 256 (512 * 26 * 26 = 346,112 input neurons to the Dense layer) (346k × 1024 = 354M+ parameters just in one FC layer) (slow training & possible overfitting)
# Better than before - 3 → 16 → 32 → 64 → 128 → 256 → 512 (Faster, deeper, but more parameters and still slower with slight overfitting)
# Best Choice - 3 → 16 → 32 → 64 → 64 → 128 → 512 (lighter, faster, lower overfitting risk, but less complex features extraction) (great balance of depth, speed, and accuracy)

import torch
import torch.nn as nn

class NumberPlateCNN(nn.Module):
    def __init__(self):
        super(NumberPlateCNN, self).__init__()

        self.features = nn.Sequential(
            # Block 1: 3 → 16
            nn.Conv2d(3, 16, kernel_size=3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),

            # Block 2: 16 → 32
            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),  # 224 → 112

            # Block 3: 32 → 64
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),

            # Block 4: 64 → 64
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),  # 112 → 56

            # Block 5: 64 → 128
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),  # 56 → 28

            # Block 6: 128 → 512
            nn.Conv2d(128, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),  # 28 → 14
        )

        self.regressor = nn.Sequential(
            nn.Flatten(),                         # Output: 512 * 14 * 14 = 100352
            nn.Linear(512 * 14 * 14, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, 4)  # [xmin, ymin, xmax, ymax]
        )

    def forward(self, x):
        x = self.features(x)
        x = self.regressor(x)
        return x
```

Create a custom Dataset class

Load:
- processed/ images
- bounding boxes from train_annotations.csv

```python
import torch
from torch.utils.data import Dataset
import cv2
import os
import pandas as pd
import numpy as np

class NumberPlateDataset(Dataset):
    def __init__(self, csv_file, image_dir, transform=None):
        self.df = pd.read_csv(csv_file)
        self.image_dir = image_dir
        self.transform = transform
        self.image_names = self.df['filename'].unique()

    def __len__(self):
        return len(self.image_names)

    def __getitem__(self, idx):
        filename = self.image_names[idx]
        img_path = os.path.join(self.image_dir, filename)
        image = cv2.imread(img_path)
        image = cv2.resize(image, (224, 224))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB).astype(np.float32) / 255.0
        image = torch.tensor(image).permute(2, 0, 1)  # [H, W, C] → [C, H, W]

        # Get all bounding boxes for this image
        boxes = self.df[self.df['filename'] == filename][['xmin', 'ymin', 'xmax', 'ymax']].values
        box = boxes[0]  # we assume one plate per image
        box = torch.tensor(box, dtype=torch.float32)

        return image, box
```

```python
def calculate_iou(pred_box, true_box):
    x1_p, y1_p, x2_p, y2_p = pred_box
    x1_t, y1_t, x2_t, y2_t = true_box

    x1_i = max(x1_p, x1_t)
    y1_i = max(y1_p, y1_t)
    x2_i = min(x2_p, x2_t)
    y2_i = min(y2_p, y2_t)

    inter_area = max(0, x2_i - x1_i) * max(0, y2_i - y1_i)
    pred_area = (x2_p - x1_p) * (y2_p - y1_p)
    true_area = (x2_t - x1_t) * (y2_t - y1_t)
    union_area = pred_area + true_area - inter_area

    return inter_area / union_area if union_area > 0 else 0.0
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm
import matplotlib.pyplot as plt

# 🛑 EarlyStopping Class
class EarlyStopping:
    def __init__(self, patience=3, min_delta=0.0):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = float('inf')
        self.early_stop = False

    def __call__(self, val_loss):
        if val_loss < self.best_loss - self.min_delta:
            self.best_loss = val_loss
            self.counter = 0
        else:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True

# Optimizer Setup
model = NumberPlateCNN().to("cuda" if torch.cuda.is_available() else "cpu")
device = next(model.parameters()).device
criterion = nn.SmoothL1Loss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

# Function to compute IoU
def calculate_iou(pred_box, true_box):
    x1_p, y1_p, x2_p, y2_p = pred_box
    x1_t, y1_t, x2_t, y2_t = true_box
    x1_i = max(x1_p, x1_t)
    y1_i = max(y1_p, y1_t)
    x2_i = min(x2_p, x2_t)
    y2_i = min(y2_p, y2_t)
    inter_area = max(0, x2_i - x1_i) * max(0, y2_i - y1_i)
    pred_area = (x2_p - x1_p) * (y2_p - y1_p)
    true_area = (x2_t - x1_t) * (y2_t - y1_t)
    union_area = pred_area + true_area - inter_area
    return inter_area / union_area if union_area > 0 else 0.0

# Training loop
train_losses, val_losses, val_ious = [], [], []
early_stopping = EarlyStopping(patience=3, min_delta=1e-4)

for epoch in range(50):  # You can set a higher limit; early stopping will cut it
    model.train()
    epoch_train_loss = 0

    for imgs, boxes in tqdm(train_loader, desc=f"Epoch {epoch+1}"):
        imgs, boxes = imgs.to(device), boxes.to(device)
        preds = model(imgs)
        loss = criterion(preds, boxes)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        epoch_train_loss += loss.item()

    train_loss = epoch_train_loss / len(train_loader)
    train_losses.append(train_loss)

    # --- Validation ---
    model.eval()
    epoch_val_loss = 0
    iou_total = 0
    with torch.no_grad():
        for imgs, boxes in val_loader:
            imgs, boxes = imgs.to(device), boxes.to(device)
            preds = model(imgs)
            loss = criterion(preds, boxes)
            epoch_val_loss += loss.item()

            for pred, true in zip(preds, boxes):
                iou = calculate_iou(pred.cpu().numpy(), true.cpu().numpy())
                iou_total += iou

    val_loss = epoch_val_loss / len(val_loader)
    mean_iou = iou_total / len(val_loader)
    val_losses.append(val_loss)
    val_ious.append(mean_iou)

    # ✅ Log results
    print(f"Epoch {epoch+1}: Train Loss = {train_loss:.4f} | Val Loss = {val_loss:.4f} | Mean IoU = {mean_iou:.4f}")

    # 📈 Plot Loss and IoU
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label="Train Loss")
    plt.plot(val_losses, label="Val Loss")
    plt.legend()
    plt.grid()
    plt.title("Loss Curve")

    plt.subplot(1, 2, 2)
    plt.plot(val_ious, label="Mean IoU", color='green')
    plt.legend()
    plt.grid()
    plt.title("Validation IoU")
    plt.show()

    # 🔒 Check early stopping
    early_stopping(val_loss)
    if early_stopping.early_stop:
        print(f"🛑 Early stopping triggered at epoch {epoch+1}")
        break
```
